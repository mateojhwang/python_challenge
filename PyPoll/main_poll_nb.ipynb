```python
import os
import csv
```

```python
election_data = os.path.join("..", "Resources", "election_data.csv")
```

```python
# Read through csv file
with open(election_data) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=",")

    # store header row
    csv_header = next(csv_file)

    #create empty lists that will be filled in a future for loop
    total_votes = []
    votes_for_cand = []
    list_of_cand = ["Charles Casper Stockham", "Diana DeGette", "Raymon Anthony Doane"]
    candidate0 = 0
    candidate1 = 0
    candidate2 = 0
    list_of_candidates = []

    # for loop that loops through the amount of rows in the csv file
    for rows in csv_reader:

        # add each row of the csv file into the list called total_votes
        total_votes.append(rows)

        # add each row[2] of the csv file into the list called votes_for_can
        votes_for_cand.append(rows[2])

    # checks each vote and adds to a new list if it is a new candidate name
    for word in votes_for_cand:

        # check if word is not in list_of_candidates
        if word not in list_of_candidates:

            # add word to new list to make a list of unique values
            list_of_candidates.append(word)
    
    # put all candidates in dictionary key and have the value count the number of votes
    vote_count_dict = {}

    # loop in votes_for_cand list
    for key in votes_for_cand:

        # check if the value is in dictionary
        if key in vote_count_dict:

            # if it is in dictionary add 1 to the value
            vote_count_dict[key] +=1
        else:
            # if it is not in dictionary add in the dictionary with value = 1
            vote_count_dict[key] = 1

    # add a for loop that will iterate until the end of the list
    for i in range(len(total_votes)):

        # compares the vote in each index to a list of the candidates names and adds 1 if voted for
        if votes_for_cand[i] == list_of_candidates[0]:
            candidate0 += 1

        elif votes_for_cand[i] == list_of_candidates[1]:
            candidate1 += 1

        else:
            candidate2 += 1


    # print everything to terminal
    print("Election Results")
    print("------------------------------")
    print(f"Total votes: {len(total_votes)}")
    print("------------------------------")
    print(f"Charles Casper Stockham: {candidate0/len(total_votes)*100} ({candidate0})")
    print(f"Diana DeGette: {candidate1/len(total_votes)*100} ({candidate1})")
    print(f"Raymon Anthony Doane: {candidate2/len(total_votes)*100} ({candidate2})")
    print("------------------------------")
    if candidate0 > candidate1 and candidate0 > candidate2:
        print(f"Winner: {list_of_cand[0]}")
    elif candidate1 > candidate0 and candidate1 > candidate2:
        print(f"Winner: {list_of_cand[1]}")
    else:
        print(f"Winner: {list_of_cand[2]}")

# print everything to output_poll.txt
with open("output_poll.txt", "w") as f:
    print("Election Results", file=f)
    print("------------------------------", file=f)
    print(f"Total votes: {len(total_votes)}", file=f)
    print("------------------------------", file=f)
    print(f"Charles Casper Stockham: {candidate0/len(total_votes)*100} ({candidate0})", file=f)
    print(f"Diana DeGette: {candidate1/len(total_votes)*100} ({candidate1})", file=f)
    print(f"Raymon Anthony Doane: {candidate2/len(total_votes)*100} ({candidate2})", file=f)
    print("------------------------------", file=f)
    if candidate0 > candidate1 and candidate0 > candidate2:
        print(f"Winner: {list_of_cand[0]}", file=f)
    elif candidate1 > candidate0 and candidate1 > candidate2:
        print(f"Winner: {list_of_cand[1]}", file=f)
    else:
        print(f"Winner: {list_of_cand[2]}", file=f)
```
